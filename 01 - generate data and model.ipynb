```python
def categorizar_programa(programa):


    programa = programa.upper()

    # Ingeniería
    if 'INGENIERIA' in programa or 'INGENIERÍA' in programa:
        return 'Ingeniería'

    # Administración y Negocios
    elif 'ADMINISTRACION' in programa or 'ADMINISTRACIÓN' in programa or 'NEGOCIOS' in programa or 'MERCADEO' in programa or 'MERCADOTECNIA' in programa or 'FINANZAS' in programa or 'COMERCIO' in programa:
        return 'Administración y Negocios'

    # Ciencias Naturales
    elif 'CIENCIAS' in programa or 'BIOLOGIA' in programa or 'BIOLÓGICA' in programa or 'QUIMICA' in programa or 'QUÍMICA' in programa or 'FISICA' in programa or 'FÍSICA' in programa:
        return 'Ciencias Naturales'

    # Matemáticas y Estadística
    elif 'MATEMATICAS' in programa or 'MATEMÁTICAS' in programa or 'ESTADISTICA' in programa or 'ESTADÍSTICA' in programa:
        return 'Matemáticas y Estadística'

    # Artes y Diseño
    elif 'ARTES' in programa or 'DISEÑO' in programa or 'DISE¿O' in programa or 'MÚSICA' in programa or 'MUSICA' in programa or 'DANZA' in programa or 'TEATRO' in programa or 'CINE' in programa or 'AUDIOVISUAL' in programa:
        return 'Artes y Diseño'

    # Educación
    elif 'LICENCIATURA' in programa or 'EDUCACION' in programa or 'EDUCACIÓN' in programa or 'PEDAGOGIA' in programa or 'PEDAGOGÍA' in programa:
        return 'Educación'

    # Ciencias Sociales y Humanidades
    elif 'CIENCIAS SOCIALES' in programa or 'HUMANIDADES' in programa or 'FILOSOFIA' in programa or 'FILOSOFÍA' in programa or 'SOCIOLOGIA' in programa or 'SOCIOLOGÍA' in programa or 'HISTORIA' in programa or 'ANTROPOLOGIA' in programa or 'ANTROPOLOGÍA' in programa or 'PSICOLOGIA' in programa or 'PSICOLOGÍA' in programa or 'DERECHO' in programa or 'CIENCIAS POLITICAS' in programa or 'CIENCIA POLITICA' in programa or 'RELACIONES INTERNACIONALES' in programa:
        return 'Ciencias Sociales y Humanidades'

    # Salud
    elif 'SALUD' in programa or 'MEDICINA' in programa or 'ENFERMERIA' in programa or 'ENFERMERÍA' in programa or 'FISIOTERAPIA' in programa or 'NUTRICION' in programa or 'NUTRICIÓN' in programa or 'ODONTOLOGIA' in programa or 'ODONTOLOGÍA' in programa or 'TERAPIA' in programa:
        return 'Salud'

    # Lenguas y Lingüística
    elif 'LINGUISTICA' in programa or 'LINGÜÍSTICA' in programa or 'IDIOMAS' in programa or 'LENGUAS' in programa or 'ESPAÑOL' in programa or 'INGLES' in programa or 'INGLÉS' in programa or 'FRANCES' in programa or 'FRANCÉS' in programa:
        return 'Lenguas y Lingüística'

    # Computación e Informática
    elif 'COMPUTACION' in programa or 'COMPUTACIÓN' in programa or 'INFORMATICA' in programa or 'INFORMÁTICA' in programa or 'SOFTWARE' in programa:
        return 'Computación e Informática'

    # Arquitectura y Construcción
    elif 'ARQUITECTURA' in programa or 'CONSTRUCCION' in programa or 'CONSTRUCCIÓN' in programa:
        return 'Arquitectura y Construcción'

    # Agricultura y Recursos Naturales
    elif 'AGRONOMIA' in programa or 'AGRONOMÍA' in programa or 'AGROFORESTAL' in programa or 'AGROINDUSTRIAL' in programa or 'ZOOTECNIA' in programa or 'VETERINARIA' in programa or 'RECURSOS NATURALES' in programa or 'AMBIENTAL' in programa or 'ECOLOGIA' in programa or 'ECOLOGÍA' in programa:
        return 'Agricultura y Recursos Naturales'

    # Comunicación y Periodismo
    elif 'COMUNICACION' in programa or 'COMUNICACIÓN' in programa or 'PERIODISMO' in programa:
        return 'Comunicación y Periodismo'

    # Turismo y Hotelería
    elif 'TURISMO' in programa or 'HOTELERIA' in programa or 'HOTELERÍA' in programa or 'GASTRONOMIA' in programa or 'GASTRONOMÍA' in programa:
        return 'Turismo y Hotelería'

    # Deportes y Actividad Física
    elif 'DEPORTE' in programa or 'ACTIVIDAD FISICA' in programa or 'ACTIVIDAD FÍSICA' in programa or 'RECREACION' in programa or 'RECREACIÓN' in programa:
        return 'Deportes y Actividad Física'

    # Seguridad y Defensa
    elif 'SEGURIDAD' in programa or 'DEFENSA' in programa or 'MILITAR' in programa or 'CRIMINALISTICA' in programa or 'CRIMINAL' in programa:
        return 'Seguridad y Defensa'

    # Bibliotecología y Archivística
    elif 'BIBLIOTECOLOGIA' in programa or 'BIBLIOTECOLOGÍA' in programa or 'ARCHIVISTICA' in programa or 'ARCHIVÍSTICA' in programa:
        return 'Bibliotecología y Archivística'

    # Ciencias Religiosas y Teología
    elif 'TEOLOGIA' in programa or 'TEOLOGÍA' in programa or 'RELIGIOSO' in programa or 'RELIGION' in programa or 'RELIGIÓN' in programa:
        return 'Ciencias Religiosas y Teología'

    # Otros
    else:
        return 'Otros'
```

```python
def clean_saberpro(train):
    # Eliminar columnas no deseadas
    train = train.drop ('FAMI_TIENEINTERNET.1', axis = 1)
    train = train.drop ('ESTU_PRIVADO_LIBERTAD', axis = 1)
    # Mapeo de valores categóricos a numéricos
    map_varios = ['FAMI_TIENEINTERNET' , 'FAMI_TIENELAVADORA', 'FAMI_TIENEAUTOMOVIL', 'ESTU_PAGOMATRICULAPROPIO', 'FAMI_TIENECOMPUTADOR']

    map = {'Si':1, 'No':0}
    for column in map_varios:
      train[column] = train[column].map(map)

    map_matricula = {
    'Menos de 500 mil': 0,
    'Entre 500 mil y menos de 1 millón': 1,
    'Entre 1 millón y menos de 2.5 millones': 2,
    'Entre 2.5 millones y menos de 4 millones': 3,
    'Entre 4 millones y menos de 5.5 millones': 4,
    'Entre 5.5 millones y menos de 7 millones': 5,
    'Más de 7 millones': 6,
    'No pago matrícula': 7
    }
    map_horas_semana = {
    '0': 0,
    'Menos de 10 horas': 1,
    'Entre 11 y 20 horas': 2,
    'Entre 21 y 30 horas': 3,
    'Más de 30 horas': 4
    }

    map_estrato = {'Sin Estrato': 0,
               'Estrato 1': 1,
               'Estrato 2': 2,
               'Estrato 3': 3,
               'Estrato 4': 4,
               'Estrato 5': 5,
               'Estrato 6': 6}

    train['ESTU_VALORMATRICULAUNIVERSIDAD'] = train['ESTU_VALORMATRICULAUNIVERSIDAD'].map(map_matricula)
    train['ESTU_HORASSEMANATRABAJA'] = train['ESTU_HORASSEMANATRABAJA'].map(map_horas_semana)
    train['FAMI_ESTRATOVIVIENDA'] = train['FAMI_ESTRATOVIVIENDA'].map(map_estrato)
    # Obtener las columnas que tienen valores faltantes

    cols_with_na = train.columns[train.isna().any()].tolist()
    for col in cols_with_na:
        # Obtener los valores no nulos de la columna
        non_na_values = train[col].dropna().values

        # Crear una máscara booleana para identificar los valores faltantes
        mask = train[col].isna()

        # Rellenar los valores faltantes con valores aleatorios de los valores existentes
        train.loc[mask, col] = np.random.choice(non_na_values, size=train[col].isna().sum())



    train['ESTU_PRGM_ACADEMICO'] = train['ESTU_PRGM_ACADEMICO'].apply(lambda x: categorizar_programa(x))
    train = pd.get_dummies(train, columns=['ESTU_VALORMATRICULAUNIVERSIDAD','ESTU_HORASSEMANATRABAJA','FAMI_ESTRATOVIVIENDA','FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE', 'ESTU_PRGM_DEPARTAMENTO','ESTU_PRGM_ACADEMICO'], prefix=['ESTU_VALORMATRICULAUNIVERSIDAD','ESTU_HORASSEMANATRABAJA','FAMI_ESTRATOVIVIENDA','PADRE_EDUCACION', 'MADRE_EDUCACION','ESTU_PRGM_DEPARTAMENTO','ESTU_PRGM_ACADEMICO'])
    train.replace({True: 1, False: 0}, inplace=True)



    return train
```

```python
dtr = train
dts = test
lentr = len(dtr)
dtr.shape, dts.shape
```

```python
source_cols = [i for i in dtr.columns if i!="RENDIMIENTO_GLOBAL"]
all_data = pd.concat((dtr[source_cols], dts[source_cols]))
all_data.index = range(len(all_data))

all_data = clean_saberpro(all_data)

Xtr, ytr = all_data.iloc[:lentr].values, dtr["RENDIMIENTO_GLOBAL"].values
Xts  = all_data.iloc[lentr:].values

print (Xtr.shape, ytr.shape)
print (Xts.shape)
```

**Train Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
# Crear el modelo DecisionTreeClassifier con max_depth=10
dt_classifier = DecisionTreeClassifier(max_depth=10)



# Entrenar el modelo con los datos de entrenamiento
dt_classifier.fit(Xtr, ytr)

# Obtener el puntaje del modelo en los datos de entrenamiento
train_score = dt_classifier.score(Xtr, ytr)
print(f"Puntaje en los datos de entrenamiento: {train_score}")

y_pred = dt_classifier.predict(Xts)

```

**Load Model and Test Data from Files**

```python
y_pred = dt_classifier.predict(Xts)

# Evaluar el modelo
accuracy = accuracy_score(dts["RENDIMIENTO_GLOBAL"], y_pred)
print(f"Precisión: {accuracy}")

f1 = f1_score(dts["RENDIMIENTO_GLOBAL"], y_pred, average='macro')
print(f"Puntaje F1: {f1}")

confusion_mat = confusion_matrix(dts["RENDIMIENTO_GLOBAL"], y_pred)
print("Matriz de confusión:")
print(confusion_mat)
```
